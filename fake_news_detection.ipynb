```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

#text = text.lower(): This line converts all characters in the text to lowercase. This step is often performed to ensure case insensitivity in further processing.

#text = re.sub('\[.*?\]', '', text): This line uses regular expressions to remove square brackets and any text within them. It is likely used to remove any references or annotations within the text.

#text = re.sub("\\W"," ",text): This line uses regular expressions to replace all non-alphanumeric characters (except underscore) with a space. This step effectively removes special characters and punctuation marks.

#text = re.sub('https?://\S+|www\.\S+', '', text): This line uses regular expressions to remove URLs or web addresses starting with "http://" or "https://" or "www.". It replaces them with an empty string.

#text = re.sub('<.*?>+', '', text): This line uses regular expressions to remove HTML tags from the text. It replaces any occurrences of opening and closing angle brackets with an empty string.

#text = re.sub('[%s]' % re.escape(string.punctuation), '', text): This line uses the string.punctuation string to remove all punctuation marks from the text. It replaces them with an empty string.

#text = re.sub('\n', '', text): This line removes newline characters from the text by replacing them with an empty string.

#text = re.sub('\w*\d\w*', '', text): This line uses regular expressions to remove any alphanumeric sequences that contain numbers. It replaces them with an empty string. This step effectively removes numbers and alphanumeric combinations with numbers.
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train , x_test, y_train, y_test= train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression() #(random_state = 0)
LR.fit(xv_train, y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)

pred_dt = DT.predict(xv_test)

DT.score(xv_test, y_test)

print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(xv_train, y_train)
```

```python
pred_gbc = GBC.predict(xv_test)

GBC.score(xv_test, y_test)

print(classification_report(y_test, pred_gbc))
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train, y_train)

```

```python
pred_rfc = RFC.predict(xv_test)

RFC.score(xv_test, y_test)

print(classification_report(y_test, pred_rfc))
```

```python
def output_lable(n):
    if n == 1:
        return "Fake News"
    elif n == 0:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)
    pred_GBC = GBC.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)

    print("\n\nLR Prediction: {}".format(output_lable(pred_LR[0])))
    print("\n\nRFC Prediction: {}".format(output_lable(pred_RFC[0])))
    print("\n\nGBC Prediction: {}".format(output_lable(pred_GBC[0])))
    print("\n\nDT Prediction: {}".format(output_lable(pred_DT[0])))
```
